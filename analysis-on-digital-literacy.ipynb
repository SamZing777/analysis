# Clean the Data

```python
# Convert Empty values to 0

data.fillna(0)
```

```python
# Remove Age Bound, Source 1, Source 2, Source 3 Columns

columns_to_remove = ["Age Bound", "Source 1", "Source 2", "Source 3"]
data.drop(columns=columns_to_remove, inplace=True)
```

```python
# Convert floats to percentage

# columns_to_convert = ["DLRate", "Android Users", "iOS Users", "Windows Users", "PC Users"]

# def float_to_percentage(data):
data["DLRate"] = data["DLRate"] * 100
data["Android Users"] = data["Android Users"] * 100
data["iOS Users"] = data["iOS Users"] * 100
data["Windows Users"] = data["Windows Users"] * 100
data["PC Users"] = data["PC Users"] * 100
data.head(5)
```

```python
# Convert data with Empty values

data.fillna(0)
```
